```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
line_data=tips.groupby("day")["total_bill"].mean()
plt.figure(figsize=(6,4))
plt.plot(line_data.index,line_data.values,marker='o',linewidth=2)
plt.title("Average of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average of total bill")
plt.grid(True)
plt.show()
```

```python
bar_data=tips.groupby("day")["tip"].mean()
plt.figure(figsize=(6,4))
plt.bar(bar_data.index,bar_data.values,color=["orange","blue","green","yellow"])
plt.title("Average day wise tip",color="green")
plt.xlabel("Day",family="sans-serif",color="red",weight="normal",size=20)
plt.ylabel("Average Tip",color="blue")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(tips["total_bill"],bins=10,color="green")
plt.title("Distribution of total Bills")
plt.xlabel("Total Bill",color="brown")
plt.ylabel("Frequency of bill",color="purple")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(tips["total_bill"],tips["tip"],color="purple")
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(tips["total_bill"])
plt.title("Total Bill on this date")
plt.xlabel("box plot of total Bill")
plt.ylabel("total bill")
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.boxplot([
    tips[tips["day"]=="Thur"]["total_bill"],
    tips[tips["day"]=="Fri"]["total_bill"],
    tips[tips["day"]=="Sat"]["total_bill"],
    tips[tips["day"]=="Sun"]["total_bill"],
],
labels=["Thur","Fri","Sat","Sun"])

plt.title("total bill by day")
plt.ylabel("total bill")
plt.show()
```

```python
corr=tips.corr(numeric_only=True)
plt.figure(figsize=(6,5))
plt.imshow(corr,cmap="coolwarm")
plt.xticks(range(len(corr.columns)),corr.columns,rotation=45)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.colorbar()
plt.title("Plot HEATMAP")

for i in range (len(corr.columns)):
    for j in range(len(corr.columns)):
        plt.text(j, i,
                 f"{corr.iloc[i,j]:.2f}",
                 ha="center",
                 va="center",
                 color="black")
plt.show()
```
